**Question 2**

How many foreign keys does the “LanguageMap” table have?

**Options:**

 - 0
 - 2
 - 3
 - 1

**Solution**

```python
%%sql 

SELECT COUNT(*)
FROM movies AS m
INNER JOIN productioncompanymap AS pm
ON pm.movie_id = m.movie_id
INNER JOIN productioncompanies AS p
ON p.production_company_id = pm.production_company_id 
WHERE
p.production_company_name = "Pixar Animation Studios"
LIMIT 10;

```

**Question 4**

What is the most popular action movie that has some German in it? (Hint: The German word for German is Deutsch)

**Options:**
 - The Bourne Identity
 - Mission: Impossible - Rogue Nation
 - Captain America: Civil War
 - Quantum of Solace

**Solution**

```python
%%sql 

SELECT m.title,MAX(m.popularity)
FROM movies AS m
INNER JOIN languagemap AS lm
ON lm.movie_id = m.movie_id
INNER JOIN languages AS l
ON lm.iso_639_1 = l.iso_639_1 
WHERE
l.language_name = "Deutsch"

LIMIT 10;
```

**Question 5**

In how many movies did Tom Cruise portray the character Ethan Hunt? (Hint: Characters are listed in the Casts table.)

**Options:**
 - 4
 - 3
 - 6
 - 5

**Solution**

```python
%%sql 

SELECT COUNT(*)
FROM movies AS m
INNER JOIN casts AS c
ON c.movie_id = m.movie_id
WHERE
c.characters = "Ethan Hunt"

LIMIT 10;
```

**Solution**

```python
%%sql 

SELECT count(*) FROM oscars as o 
INNER JOIN movies AS m
ON m.title = o.name
WHERE o.award ="Best Picture" 
limit 10;
```

**Question 8** 

How many movies contain at least one of the languages, Afrikaans or Zulu?

**Options:**
 - 10
 - 8
 - 12
 - 15

**Solution**

```python
%%sql 
```

```python
%%sql 
SELECT COUNT(1) FROM(
SELECT m.movie_id
FROM movies AS m
INNER JOIN languagemap AS lm
ON lm.movie_id = m.movie_id
INNER JOIN languages AS l
ON lm.iso_639_1 = l.iso_639_1 
WHERE
l.language_name = "Afrikaans" OR  l.language_name = "isiZulu"
GROUP BY  m.movie_id
    HAVING COUNT(DISTINCT l.language_name)>0
) as g ;
```

**Question 9**

In which country was the movie “Star Wars” produced?  

**Options:**
 - Canada
 - United Kingdom
 - France
 - United States of America

**Solution**

```python
%%sql 


SELECT p.production_country_name
FROM movies AS m
INNER JOIN productioncountrymap AS pc
ON pc.movie_id = m.movie_id
INNER JOIN productioncountries AS p
ON p.iso_3166_1 = pc.iso_3166_1 
WHERE
m.title= "Star Wars"
LIMIT 10;
```

**Question 10**

How many movies are in the database that are both a Romance and a Comedy?

**Options:**

 - 373
 - 484
 - 262
 - 595

**Solution**

```python
%%sql 
SELECT COUNT(1) FROM(
SELECT m.movie_id
FROM movies AS m
INNER JOIN genremap AS gm
ON gm.movie_id = m.movie_id
INNER JOIN genres AS g
ON g.genre_id = gm.genre_id
WHERE
 g.genre_name in ("Romance","Comedy")
GROUP BY m.movie_id
HAVING COUNT(DISTINCT g.genre_name)= 2)
as t
```
