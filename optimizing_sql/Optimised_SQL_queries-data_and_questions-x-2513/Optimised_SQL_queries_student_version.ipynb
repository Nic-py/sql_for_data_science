**Question 5**

Which day of the week had the highest number of cancelled flights?(Note 1 = cancelled , 0 = not cancelled;)


**Options:**
- Monday(1)
- Wednesday(3) 
- Saturday(6) 
- Thursday (4) 

```python
%%sql
WITH CTE AS
(SELECT DayOfWeek, COUNT(Cancelled) AS NoCancelled
FROM flights
WHERE Cancelled = 1
GROUP BY DayOfWeek)
SELECT DayOfWeek, MAX(NoCancelled) FROM CTE
```

**Question 7**

What is the most produced model for the manufacturer "BOEING"?

**Options:**
- 717-200
- 737-7H4
- 757-222
- 737-3H4

```python
%%sql 
WITH CTE AS
(SELECT manufacturer, model, COUNT(model) AS noProduced
FROM planes
WHERE manufacturer = "BOEING"
GROUP BY model)

SELECT manufacturer, model, MAX(noProduced) FROM CTE  
```

**Question 8**

What manufacturer had the highest average delay time (DepDelay + ArrDelay)? 

**Options:**
- PAIR MIKE E
- AERONCA
- AVIAT AIRCRAFT INC
- BOEING OF CANADA LTD

```python
%%sql
WITH CTE AS
    (SELECT f.DepDelay, f.ArrDelay, (f.DepDelay + f.ArrDelay) AS averageDelay, p.manufacturer
    FROM flights AS f
    INNER JOIN planes AS p
    ON f.TailNum = p.tailnum)
SELECT DepDelay, ArrDelay, manufacturer, averageDelay
FROM CTE 
WHERE manufacturer = "PAIR MIKE E" 
    OR manufacturer = "AERONCA"
    OR manufacturer = "AVIAT AIRCRAFT INC"
    OR manufacturer = "BOEING OF CANADA LTD"
ORDER BY averageDelay DESC

```

**Question 9**

How many planes landed at Los Angeles International Airport?

```python
%%sql 
SELECT COUNT(*) FROM airports as a 
inner join flights as f
on a.iata = f.Origin
where airport = "Los Angeles International" 
```

**Question 10**

Which domestic carrier had the best on-time performance (OTP), where OTP is defined as the rate of on time flights with a 15min buffer on departure and arrival?

**Options:**
- American Airlines Inc
- Hawaiian Airlines Inc
- Comair Inc
- Aloha Airlines Inc

```python
%%sql
with cte as
(select c.Code, c.Description, f.DepTime, f.ArrTime, f.ArrDelay, f.DepDelay from carriers as c
left join flights as f
on c.Code = f.UniqueCarrier)
select * from cte where DepTime is not null and Description like "%American%"
limit 200
```
