1. What query would you write to change the name of the language with the 'zh' iso code in the "Languages" table to Chinese?

```python
%%read_sql

UPDATE languages
SET language_name = "Chinese"
WHERE iso_639_1 = "zh"
```

```python
%%read_sql

SELECT * 
FROM languages 
WHERE iso_639_1 = "zh"
LIMIT 10
```
